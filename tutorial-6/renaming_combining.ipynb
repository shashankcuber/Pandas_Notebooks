```python
import pandas as pd
pd.set_option('display.max_rows', 5)
reviews = pd.read_csv("../Dataset/winemag-data-130k-v2.csv", index_col = 0)
reviews
```

### Renaming

To rename a column or index we use the function rename()

```python
reviews.rename(columns = {'points': 'score'})
```

Rename even allows to rename the index or columns keyword parameter. It supports a variety of input formats but dictionary is very convenient

```python
reviews.rename(index = {0:'first', 1:'second'})
```

Renaming index is very rare mostly we rename columns and when we want to rename index set_index() is used often. 
Both index of rows and columns can be changed and this can be done through rename_axis().

```python
reviews.rename_axis("wines", axis="rows").rename_axis("fields", axis='columns')
```

### Combining 
This deals with combining dataframes. 
It is done through three core methods in pandas: concat(), join(), merge(). Mostly the operations of merge can be done through join() as well. 
The simplest one is concat(). This basically glue dfs along a specified axis.

```python
canadian_youtube = pd.read_csv("../Dataset/CAvideos.csv")
british_youtube = pd.read_csv("../Dataset/GBvideos.csv")
```

```python
pd.concat([canadian_youtube, british_youtube])
```

The better form of combining is done through join, it gives all flexiblities like joins in SQL for tables. 
It helps to combine DFs that have a common index.

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])
```

```python
left.join(right, lsuffix = '_CAN', rsuffix = '_UK')
```

The lsuffix and rsuffix is a really cool way to distinguish the two dfs columns when both the dfs share a common column names.
